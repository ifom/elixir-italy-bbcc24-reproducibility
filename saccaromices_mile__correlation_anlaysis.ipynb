```python
# LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# INPUT
mile_file = "saccharomyces_mile.csv"
```

```python
# PROCESSING
## Import dataset
data = pd.read_csv( mile_file, sep = ';' )
data.head()
```

```python
## Convert Growth column into float (Excel has setted ',' as integer-decimal separator, python want '.')
data[ "Growth" ] = data[ "Growth" ].str.replace( ',', '.' ).astype( "float" )
```

```python
# OUPTUT
## Plot dataset
data.plot.scatter( x = "Time", y = "Growth" )
plt.savefig( "Smile.pdf" )
```

```python
## Perform correlation
corr_val = np.corrcoef( data[ "Time" ], data[ "Growth" ] )[ 0, 1 ]
print( corr_val )
```
